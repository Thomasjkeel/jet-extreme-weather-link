```python
def calc_perc_coverage(row, land_cells):
    return (row.sum() / land_cells)*100
```

```python
%%time
# ~45 seconds
extreme_low_temp_sub_reduced['wna_perc_coverage'] = extreme_low_temp_sub_reduced.where(ar6_land_mask == 3)['t2m_detrended'].groupby('time').apply(calc_perc_coverage, args=(NUM_WNA_CELLS,))
```

```python
%%time
# ~45 seconds
extreme_low_temp_sub_reduced['cna_perc_coverage'] = extreme_low_temp_sub_reduced.where(ar6_land_mask == 4)['t2m_detrended'].groupby('time').apply(calc_perc_coverage, args=(NUM_CNA_CELLS,))
```

```python
%%time
# ~45 seconds
extreme_low_temp_sub_reduced['ena_perc_coverage'] = extreme_low_temp_sub_reduced.where(ar6_land_mask == 5)['t2m_detrended'].groupby('time').apply(calc_perc_coverage, args=(NUM_ENA_CELLS,))
```

### percentiles

```python
## Because the 95th percentile can be like 2% coverage for July, I am using a minimum coverage percentage
MIN_COVERAGE_PERC = 5.0
```

```python
COVERAGE_PERCENTILE = 0.95
```

```python
wna_month_percentiles = {}
cna_month_percentiles = {}
ena_month_percentiles = {}
for month in np.unique(extreme_low_temp_sub_reduced.time.dt.month):
    month_data = extreme_low_temp_sub_reduced.sel(time=extreme_low_temp_sub_reduced.time.dt.month.isin([month]))
    wna_month_percentiles[month] = max(float(month_data['wna_perc_coverage'].quantile([COVERAGE_PERCENTILE]).data), MIN_COVERAGE_PERC)
    cna_month_percentiles[month] = max(float(month_data['cna_perc_coverage'].quantile([COVERAGE_PERCENTILE]).data), MIN_COVERAGE_PERC)
    ena_month_percentiles[month] = max(float(month_data['ena_perc_coverage'].quantile([COVERAGE_PERCENTILE]).data), MIN_COVERAGE_PERC)
```

```python
def above_month_95perc(row, region_percentiles):
    month = int(row.time.dt.month)
    return (row > region_percentiles[month]).astype(int)
```

```python
%%time
extreme_low_temp_sub_reduced['cold_air_event_WNA'] = extreme_low_temp_sub_reduced['wna_perc_coverage'].groupby('time').apply(lambda row: above_month_95perc(row, wna_month_percentiles))
```

```python
%%time
extreme_low_temp_sub_reduced['cold_air_event_CNA'] = extreme_low_temp_sub_reduced['cna_perc_coverage'].groupby('time').apply(lambda row: above_month_95perc(row, cna_month_percentiles))
```

```python
%%time
extreme_low_temp_sub_reduced['cold_air_event_ENA'] = extreme_low_temp_sub_reduced['ena_perc_coverage'].groupby('time').apply(lambda row: above_month_95perc(row, ena_month_percentiles))
```

```python
def count_unique_combinations(array_2d):
    unique_collections, counts = np.unique(array_2d, axis=0, return_counts=True)
    unique_counts = dict(zip(map(tuple, unique_collections), counts))
    
    return unique_counts


def plot_histogram_of_unique_combinations(unique_counts):
    labels, values = zip(*unique_counts.items())
    
    plt.bar(range(len(labels)), values, tick_label=list(map(str, labels)))
    plt.xlabel('Unique Collections')
    plt.ylabel('Count')
    plt.title('Histogram of Unique Collections')
    # plt.show()
```

```python
cold_air_event_combinations = extreme_low_temp_sub_reduced['cold_air_event_stacked'].T
combinations = count_unique_combinations(cold_air_event_combinations)
print(combinations)
combinations_only_events = combinations
combinations_only_events.pop(list(combinations_only_events.keys())[0])
plot_histogram_of_unique_combinations(combinations_only_events)
# plt.savefig('figures/comparison_of_unique_combinations_q3_15percthresh.png')
```

```python
def isolate_clusters(arr, min_cluster_size, fill_gaps_of_one=False):
    """
    adjusted from chatGPT
    """
    # Create a binary array where values that meet the condition are True
    clustered_values = arr > 0  # Replace this condition with your own
    if fill_gaps_of_one:
        clustered_values = make_fill_gaps_of_one(clustered_values)
    # Label connected components
    labeled_clusters, num_clusters = scipy.ndimage.label(clustered_values)

    # Count the size of each cluster
    cluster_sizes = np.bincount(labeled_clusters.ravel())

    # Mask clusters based on size condition
    valid_clusters_mask = cluster_sizes[labeled_clusters] >= min_cluster_size
    result_array = arr.copy()
    result_array[~valid_clusters_mask] = 0 # removed as event

    return result_array


def make_fill_gaps_of_one(arr):
    filled_array = arr.copy()

    for i in range(1, len(arr) - 1):
        if arr[i - 1] == 1 and arr[i + 1] == 1 and arr[i] == 0:
            filled_array[i] = 1

    return filled_array


```

```python
# Example usage:
arr = np.array([0, 1, 1, 0, 1, 0, 0, 0, 1, 0, 1])
result = isolate_clusters(arr, min_cluster_size=3)
print(result)
result = isolate_clusters(arr, min_cluster_size=3, fill_gaps_of_one=True)
print(result)
```
